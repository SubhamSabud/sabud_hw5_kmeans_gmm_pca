```python
import numpy as np
import matplotlib.pyplot as plt

# Define the mean and covariance matrices for each distribution
mean_a = [-1, -1]
mean_b = [-1, 1]
mean_c = [0, 1]
cov_a = [[2, 0.5], [0.5, 1]]
cov_b = [[1, -0.5], [-0.5, 2]]
cov_c = [[0, 1], [1, 2]]

# Define the values of sigma
sigma_values = [0.5, 1, 2, 4, 8]

# Create a figure with five subplots
fig, axes = plt.subplots (1, 5, figsize=(15, 3), sharex=True, sharey=True)

# Loop over the values of sigma
for i, sigma in enumerate (sigma_values):
    # Scale the covariance matrices by sigma^2
    cov_a_scaled = np.multiply (cov_a, sigma)
    cov_b_scaled = np.multiply (cov_b, sigma)
    cov_c_scaled = np.multiply (cov_c, sigma)
    
    # Generate 100 points from each distribution
    x_a, y_a = np.random.multivariate_normal (mean_a, cov_a_scaled, 100).T
    x_b, y_b = np.random.multivariate_normal (mean_b, cov_b_scaled, 100).T
    x_c, y_c = np.random.multivariate_normal (mean_c, cov_c_scaled, 100).T
    
    # Plot the points on the i-th subplot
    axes [i].scatter (x_a, y_a, color='red', label='Pa')
    axes [i].scatter (x_b, y_b, color='blue', label='Pb')
    axes [i].scatter (x_c, y_c, color='green', label='Pc')
    axes [i].set_title (f'sigma = {sigma}')
    axes [i].set_xlim (-20, 20)
    axes [i].set_ylim (-20, 20)
    
# Add a legend and show the figure
fig.legend (loc='upper center', ncol=3)
plt.show ()
```
